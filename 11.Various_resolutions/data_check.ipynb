```python
import pandas as pd
```

```python
df1 = pd.read_csv('data/df_final.csv')
df2 = pd.read_csv('data/no_df.csv')
```

```python
df1 = df1.set_index(['picture_name', 'handedness'])
df2 = df2.set_index(['picture_name', 'handedness'])
```

```python
merged = df1.merge(df2, left_index=True, right_index=True, suffixes=('_x', '_y'), how='outer')
c = 'hand_min_x'
diff_rows = merged.loc[(merged[c +'_x'] != merged[c +'_y']).fillna(False)]
```

```python
merged_df = pd.merge(df1, df2, on=['picture_name', 'handedness'], how='outer', indicator=True)
```

```python
# get the rows that belong to df1
df1_rows = merged_df[merged_df['_merge'] == 'left_only']

# get the rows that belong to df2
df2_rows = merged_df[merged_df['_merge'] == 'right_only']

# print the resulting DataFrames
print(df1_rows)
print(df2_rows)
```

```python
both_df = merged_df[merged_df['_merge'] == 'both']

new_list = [s + '_y' for s in df2.columns]

# check if the values in each row are the same
diff_rows = both_df.loc[:, df1.columns].ne(both_df.loc[:, new_list]).any(axis=1)

# print the resulting DataFrame
print(diff_rows)
```
